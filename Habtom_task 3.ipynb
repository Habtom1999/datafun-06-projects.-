What state has more covid case per 100000

```python
highest_cases_State = df.loc[df['RatePer100000'].idxmax(), 'State']
```

```python
print("State with the highest number of COVID cases per 100,000 people:",highest_cases_State)
```

```python
lowest_cases_State = df.loc[df['RatePer100000'].idxmin(), 'State']
```

```python
print("State with the lowest number of COVID cases per 100,000 people:",lowest_cases_State)
```
